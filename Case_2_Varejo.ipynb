```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('/content/varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique().reset_index()
#agrupar por canal de vendas
#nunique - conta quantidade de linhas distintas de idcompra
```

```python
#subtituir app para aplicativo
vendas['idcanalvenda'] = vendas['idcanalvenda'].str.replace('APP', 'Aplicativo')
vendas.groupby("idcanalvenda").idcompra.nunique().reset_index()
```

```python
vendas.groupby("bandeira").idcompra.nunique().reset_index()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique().reset_index()
```

```python
vendas.groupby("estado").idcompra.nunique().reset_index()
```

```python
vendas['Nome_Departamento'] = vendas['Nome_Departamento'].str.replace(' ', '_')
vendas.groupby("Nome_Departamento").idcompra.nunique().reset_index()
```

```python
vendas.info()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas_sem_nulos = vendas.dropna()
vendas_sem_nulos.info()
```

```python
vendas['estado'].fillna('MS', inplace = True) #inplace - quero manter isso na base original
#vendas['estado'] = vendas['estado'].fillna('MS') - antes fazia desse jeito
```

```python
vendas.info() #só faltou o preço
```

```python
#substituir os valores nulos pela media de todas as outras
media_preco = vendas['Preço'].mean() #mean = média

vendas['Preço'].fillna(media_preco, inplace = True)
```

```python
media_preco
```

```python
vendas.info() #agora todos os valores não estão nulos
```

```python
vendas[["Preço", 'Preço_com_frete']].describe()
```

```python
preco_errado = vendas.query("Preço > Preço_com_frete")

vendas_correto = vendas.query("Preço < Preço_com_frete ")
```

```python
vendas_correto.describe()
```

```python
vendas_correto.head()
```

```python
vendas_correto.query("Nome_Departamento == 'Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
vendas_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending = False).reset_index()
```

## AGRUPAR O DEPARTAMENTO E AGREGAR PELA MÉDIA DE PREÇO


```python
round(vendas_correto.groupby('Nome_Departamento')["Preço_com_frete"].agg('mean').sort_values(ascending = False).reset_index(), 2) #arrendondar duas casas decimais
```

```python
vendas_correto.groupby('Data').idcompra.nunique().sort_values(ascending = False)
```

```python
#adicionar coluna de mes para depois agrupar os valores por mes
vendas_correto['mes'] = vendas_correto["Data"].dt.month #dateTime mes
vendas_correto.head()
```

```python
vendas_correto.groupby('mes').idcompra.nunique().sort_values(ascending = False)
```

```python
cliente = pd.read_excel('cliente_varejo.xlsx')
cliente.head()
```

```python
vendas_correto.head()
#tem a cliente_log para as duas: para fazer join
```

```python
cliente[["idade","renda"]].describe()
```

```python
cliente.info() #não há dados nulos
```

```python
cliente = cliente.astype({'renda': 'float'})
cliente.info()
```

```python
vendas_cliente = vendas_correto.merge(cliente, how = 'left', on = 'cliente_Log')
vendas_cliente.head()
```

```python
agg_idcanal_renda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending = False).reset_index(), 2)
agg_idcanal_renda
```

```python
agg_idade_bandeira = round(vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending = False).reset_index(),2)
agg_idade_bandeira
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize = (7,4))
           #eixo X                            #eixo y
plt.bar(agg_idade_bandeira['bandeira'], agg_idade_bandeira['idade'], color = 'green')

plt.ylabel('Média de idade')

plt.title("Idade média por Bandeira", loc="center", fontsize=18, fontweight = 'bold', color = 'black')
```

```python
fig = plt.figure(figsize = (7,4))

plt.bar(agg_idcanal_renda['idcanalvenda'], agg_idcanal_renda['renda'], color = 'green')

plt.ylabel('Renda média')

plt.xlabel('Canal de venda')

plt.title('Renda média por canal de venda', loc = 'center', fontsize = 18, fontweight='bold', color = 'black')
```

```python
vendas_por_data = vendas_correto.groupby('Data').idcompra.nunique().reset_index()
vendas_por_data
```

```python
fig = plt.figure(figsize = (20,4))

plt.xlabel("Data da venda")

plt.ylabel("Quantidade de vendas")

x= vendas_por_data["Data"]
y = vendas_por_data["idcompra"]

plt.plot(x,y,color = 'green', linewidth = '2') #grossura da linha
plt.show()
```

```python
agg_dept_preco = round(vendas_correto.groupby('Nome_Departamento')["Preço_com_frete"].agg('mean').sort_values(ascending = False).reset_index(), 2)
agg_dept_preco
```

```python
fig = plt.figure(figsize = (35,8))

plt.bar(agg_dept_preco['Nome_Departamento'], agg_dept_preco['Preço_com_frete'], color = 'green')

plt.ylabel('Preço médio')

plt.xlabel('Departamento')

plt.title('Média de preço por departamento', loc = 'center', fontsize = 18, fontweight='bold', color = 'black')
```
